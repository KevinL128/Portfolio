# 1. Understanding the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Escritorio\Kaggle Data\Traffic_Crashes_-_Crashes.csv")

df
```

```python
# This calculate the percentage of missing values in each column of the dataset
prct_na = (df.isna().sum() / len(df)) * 100
prct_na.sort_values(ascending = False)
```

# 2. Data Preparation

```python
## We drop columns with high percentage of null value 
df = df.drop(columns =['WORKERS_PRESENT_I','DOORING_I','WORK_ZONE_TYPE', 'WORK_ZONE_I', 'PHOTOS_TAKEN_I','STATEMENTS_TAKEN_I', 
                'NOT_RIGHT_OF_WAY_I', 'CRASH_DATE_EST_I', 'LANE_CNT','INTERSECTION_RELATED_I', 'HIT_AND_RUN_I']) 
```

```python
## We drop the others na values to clean the data
df = df.dropna()
```

```python
# After checking for missing values in the dataset, we confirmed that there are non, as all columns have a count of zero for null values
df.isna().sum()
```

```python
#We change the data type of CRASH_DATE from object to time
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
```

# 3. Univariate Analysis


```python
#Crash severity

df['MOST_SEVERE_INJURY'].value_counts()
```

```python
df['MOST_SEVERE_INJURY'].value_counts().plot(kind ='bar', color= 'firebrick')

plt.xlabel("Injury Type")
plt.ylabel("Count")
plt.title("Most Severe Injury")
plt.xticks(rotation=45)

plt.show()
```

```python
#Crash cause
df['WEATHER_CONDITION'].value_counts()
```

```python
df['WEATHER_CONDITION'].value_counts().head(6).plot(kind ='bar', color= 'teal')

plt.xlabel("Weather Type")
plt.ylabel("Count")
plt.title("Weather Condition")
plt.xticks(rotation=45)

plt.show()
```

```python
#Crash cause
df['LIGHTING_CONDITION'].value_counts()
```

```python
df['LIGHTING_CONDITION'].value_counts().plot(kind ='bar', color= 'goldenrod')

plt.xlabel("Lighting Type")
plt.ylabel("Count")
plt.title("Lighting Condition")
plt.xticks(rotation=45)

plt.show()
```

```python
#Crash in day of the week
day_map = {1: 'Monday', 
           2: 'Tuesday', 
           3: 'Wednesday', 
           4: 'Thursday', 
           5: 'Friday', 
           6: 'Saturday', 
           7: 'Sunday'}

day_count = df['CRASH_DAY_OF_WEEK'].value_counts().sort_index()

day_count.index = day_count.index.map(day_map)
day_count
```

```python
day_count.plot(kind ='bar', color= 'yellowgreen')

plt.xlabel("Day of the week")
plt.ylabel("Count")
plt.title("Crash day of the week")
plt.xticks(rotation=45)

plt.show()
```

```python
#Crash in month of the year 
month_map = {1:'January', 
           2: 'February', 
           3: 'March', 
           4: 'April', 
           5: 'May', 
           6: 'June', 
           7: 'July',
           8: 'August', 
           9: 'September', 
           10: 'October', 
           11: 'November', 
           12: 'December'}

month_count = df['CRASH_MONTH'].value_counts().sort_index()

month_count.index = month_count.index.map(month_map)
month_count
```

```python
month_count.plot(kind ='bar', color= 'darkorchid', figsize = (10,6))

plt.xlabel("Month of the year")
plt.ylabel("Count")
plt.title("Month Crash")
plt.xticks(rotation=45)

plt.show()
```

# 3.1. Bivariate Analysis


```python
#Generate a bar chart displaying the severity of crashes by month 

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']


crash_severity_by_month = pd.crosstab(df['CRASH_MONTH'].map(month_map), df['MOST_SEVERE_INJURY'])

crash_severity_by_month = crash_severity_by_month.reindex(month_order)


crash_severity_by_month.plot(kind='bar', stacked=True, figsize=(10, 6))


plt.xlabel("Day of the Week")
plt.ylabel("Number of Crashes")
plt.title("Crash Severity by Day of the Week")
plt.xticks(rotation=45)
plt.legend(title="Most Severe Injury")

plt.show()

```

```python
# Heatmap visualization of weather condition and most severe injury

weather_severity = pd.crosstab(df['WEATHER_CONDITION'], df['MOST_SEVERE_INJURY'])

plt.figure(figsize=(12, 6))
sns.heatmap(weather_severity, annot=True, cmap="Blues", fmt="d")


plt.xlabel("Most Severe Injury")
plt.ylabel("Weather Condition")
plt.title("Crash Severity by Weather Condition")
plt.xticks(rotation=20)

plt.show()
```
